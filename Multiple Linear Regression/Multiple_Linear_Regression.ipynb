# 0. Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.linear_model import LinearRegression
from numpy import mean
from numpy import absolute
from numpy import sqrt
```

# 1.1 Importing the train dataset

```python
train_dataset = pd.read_csv('train.csv')
train_dataset
```

# 1.2 Removing unwanted white spaces

```python
train_dataset['col_0'] = train_dataset["col_0"].apply(lambda x: str(x).replace(' ', ''))
train_dataset['col_1'] = train_dataset["col_1"].apply(lambda x: str(x).replace(' ', ''))
train_dataset['col_2'] = train_dataset["col_2"].apply(lambda x: str(x).replace(' ', ''))
train_dataset['col_3'] = train_dataset["col_3"].apply(lambda x: str(x).replace(' ', ''))
train_dataset['col_5'] = train_dataset["col_5"].apply(lambda x: str(x).replace(' ', ''))
train_dataset['col_6'] = train_dataset["col_6"].apply(lambda x: str(x).replace(' ', ''))
train_dataset
```

# 1.3 Replacing typing mistakes

```python
train_dataset['col_0'] = train_dataset["col_0"].apply(lambda x: str(x).replace('O', '0'))
train_dataset['col_1'] = train_dataset["col_1"].apply(lambda x: str(x).replace('O', '0'))
train_dataset['col_2'] = train_dataset["col_2"].apply(lambda x: str(x).replace('O', '0'))
train_dataset['col_3'] = train_dataset["col_3"].apply(lambda x: str(x).replace('O', '0'))
train_dataset['col_5'] = train_dataset["col_5"].apply(lambda x: str(x).replace('O', '0'))
train_dataset['col_6'] = train_dataset["col_6"].apply(lambda x: str(x).replace('O', '0'))
train_dataset
```

# 1.4 Removing alphabetic characters

```python
train_dataset['col_0'] = train_dataset["col_0"].str.extract(pat='(\d+)', expand=False)
train_dataset['col_1'] = train_dataset["col_1"].str.extract(pat='(\d+)', expand=False)
train_dataset['col_2'] = train_dataset["col_2"].str.extract(pat='(\d+)', expand=False)
train_dataset['col_3'] = train_dataset["col_3"].str.extract(pat='(\d+)', expand=False)
train_dataset['col_5'] = train_dataset["col_5"].str.extract(pat='(\d+)', expand=False)
train_dataset['col_6'] = train_dataset["col_6"].str.extract(pat='(\d+)', expand=False)
train_dataset
```

# 1.5 Checking if null values present

```python
train_dataset.isna().sum()
```

# 1.6 Replacing null values with most frequent values

```python
train_dataset = train_dataset.fillna(train_dataset.mode().iloc[0])
train_dataset
```

```python
train_dataset.isna().sum()
```

# 1.7 Checking dataset info

```python
train_dataset.info()
```

# 1.8 Converting all data types into int

```python
train_dataset['col_0'] = train_dataset["col_0"].astype(int)
train_dataset['col_1'] = train_dataset["col_1"].astype(int)
train_dataset['col_2'] = train_dataset["col_2"].astype(int)
train_dataset['col_3'] = train_dataset["col_3"].astype(int)
train_dataset['col_5'] = train_dataset["col_5"].astype(int)
train_dataset['col_6'] = train_dataset["col_6"].astype(int)
```

```python
#define predictor and response variables
X = train_dataset[['col_0', 'col_1','col_2', 'col_3','col_4', 'col_5','col_6']]
y = train_dataset['y']
```

```python
X
```

```python
y
```

# 1.1.0 Build and fit model

```python
lm = LinearRegression()

# Fit our training data
model = lm.fit(X, y)
```

```python
print('Intercept: \n', model.intercept_)
print('Coefficients: \n', model.coef_)
```

# 1.1.1 K fold cross validation

```python
# create a KFold object with 5 splits 
folds = KFold(n_splits = 5, shuffle = True, random_state = 100)
scores = cross_val_score(lm, X, y, scoring='r2', cv=folds)
scores   
```

# 1.1.2 Evaluate RMSE

```python
sqrt(mean(absolute(scores)))
```

# 1.1.3 Custome Prediction

```python
custome_prediction = int(lm.predict([[0,0,11,1,100,1,2]]))

print(custome_prediction)
```
